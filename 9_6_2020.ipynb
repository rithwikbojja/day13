```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```
a = np.arange(1,11)
b = np.arange(11,21)
plt.xlabel("a values")
plt.ylabel("b values")
plt.title("a vesus b")
plt.plot(a,b,color="orange",lw=5,linestyle="-.")#lw or linewidth
plt.show()
```

```
days = list(range(1,8))
temp = [34,36,40,41,42,34,44]
plt.xlabel("days values")
plt.ylabel("temp values")
plt.title("days vesus temp")##can use xlim,ylim 
plt.plot(days,temp,color="blue",lw=3,linestyle="dashed")#lw or linewidth
plt.show()
```

```
x=np.arange(1,10)
plt.plot(x,x,label="x vlues")
plt.plot(x,2*x,label="2x vlues")
plt.plot(x,3*x,label="3x vlues")
plt.plot(x,4*x,label="4x vlues")
plt.legend()
plt.show()
```

```
x = np.linspace(0, 5, 10)
y = np.linspace(3, 6, 10)
plt.plot(x,y,"r.",x,y*2,"b+",x,y*3,"g^")
plt.show()
```

```
sub=['C', 'C++', 'Java', 'Python', 'PHP'] 
students=[23,17,35,29,12]
preference = [90,85,90,95,80]
plt.xlabel("lanf")
plt.ylabel("no of stu")
plt.bar(sub,students)
```

```
xpos=np.arange(len(sub))
xpos
```

```
plt.bar(xpos-0.2,students,label="students",width=0.4)
plt.bar(xpos+0.2,preference,label="preferences",width=0.4)
plt.legend(loc="best")
plt.xticks(xpos,sub)
plt.show()
```
